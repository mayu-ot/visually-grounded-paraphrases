```python

def write_phrase_pair_wt_gtroi(split):
    anno_base = '../../Dataset/Flickr30kEntities/Annotations/'
    sent_base = '../../Dataset/Flickr30kEntities/Sentences/'
    phrase_data = pd.read_csv('data/pl-clc/phrase_pair_%s.csv' % split)

    roi1 = []
    roi2 = []

    fp = IntProgress(min=0, max=len(phrase_data))
    display(fp)
    
    cur_id = 0
    
    for i, row in phrase_data.iterrows():
        if row['image'] != cur_id:
            bbox_list = load_bbox(anno_base+'%i.xml'% row['image'])
            entity_list = load_entity(sent_base + '%i.txt' % row['image'])
            cur_id = row['image']

        for p_k in ['org_phrase1', 'org_phrase2']:
            org_phr = row[p_k]

            ids = [k for k, v in entity_list.items() if org_phr in v]

            cords = []
            for e_id in ids:
                cords += [cord for cord, name in bbox_list if e_id in name]

            # marge bbox
            cords = np.asarray(cords)
            y_min, x_min = cords[:, :2].min(axis=0)
            y_max, x_max = cords[:, -2:].max(axis=0)

            if p_k == 'org_phrase1':
                roi1.append((y_min, x_min, y_max, x_max))
            else:
                roi2.append((y_min, x_min, y_max, x_max))
        
        fp.value = i

    phrase_data['roi1'] = roi1
    phrase_data['roi2'] = roi2
    phrase_data.to_csv('data/gt-roi/phrase_pair_%s.csv' % split)
#         for cord in cords:
#             y_min, x_min, y_max, x_max = cord
#             data.append({'image': row['image'], 'org_phrase': org_phr, 'phrase':row['phrase'], 'xmin': x_min, 'ymin': y_min, 'xmax': x_max, 'ymax': y_max})
```

```python
write_phrase_pair_wt_gtroi('train')
```
