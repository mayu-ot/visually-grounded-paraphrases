```python
import pandas as pd
import numpy as np
```

```python
from ipywidgets import IntProgress
from IPython.display import display
```

```python
def save_feature_indices_gtroi(split):
    phrase_data = pd.read_csv('data/gt-roi/phrase_pair_%s.csv' % split)
    roi_meta = pd.read_csv('data/region_feat/gt-roi-frcnn/meta_%s.csv' % split)
    N = len(phrase_data)
    feat_indices = np.zeros((N, 2), dtype=np.int32)
    
#     ip = IntProgress(min=0, max=len(phrase_data))
#     display(ip)
    
    search_cache = {}
    
    for i, row in phrase_data.iterrows():
        img = row['image']
        roi1, roi2 = row['roi1'], row['roi2']
        
        key1 = '%i-%s' % (img, roi1)
        key2 = '%i-%s' % (img, roi2)
        
        if key1 in search_cache:
            idx1 = search_cache[key1]
        else: 
            idx1 = roi_meta.query("image == %i & roi == '%s'" % (img, roi1))['Unnamed: 0']
            search_cache[key1] = idx1
        
        if key2 in search_cache:
            idx2 = search_cache[key2]
        else: 
            idx2 = roi_meta.query("image == %i & roi == '%s'" % (img, roi2))['Unnamed: 0']
            search_cache[key2] = idx2
        
        feat_indices[i, :] = idx1, idx2
        
        if i % 10000 == 0:
            print('%i / %i' % (i, N))
        
#         ip.value=i
    np.save('data/gt-roi/gt-roi-frcnn_indices_%s' % split, feat_indices)

def save_feature_indices_plclcroi(split):
    phrase_data = pd.read_csv('data/pl-clc/phrase_pair_wt_plclcbbox_%s.csv' % split)
    roi_meta = pd.read_csv('data/region_feat/pl-clc-frcnn/meta_%s.csv' % split)
    N = len(phrase_data)
    feat_indices = np.zeros((N, 2), dtype=np.int32)
    
#     ip = IntProgress(min=0, max=len(phrase_data))
#     display(ip)
    
    search_cache = {}
    
    for i, row in phrase_data.iterrows():
        img = row['image']
        roi1, roi2 = row['roi1'], row['roi2']
        
        key1 = '%i-%i' % (img, roi1)
        key2 = '%i-%i' % (img, roi2)
        
        if key1 in search_cache:
            idx1 = search_cache[key1]
        else: 
            idx1 = roi_meta.query("image == %i & roi == %i" % (img, roi1))['Unnamed: 0']
            search_cache[key1] = idx1
        
        if key2 in search_cache:
            idx2 = search_cache[key2]
        else: 
            idx2 = roi_meta.query("image == %i & roi == %i" % (img, roi2))['Unnamed: 0']
            search_cache[key2] = idx2
        
        feat_indices[i, :] = idx1, idx2
        
        if i % 10000 == 0:
            print('%i / %i' % (i, N))
        
#         ip.value=i
    np.save('data/pl-clc/pl-clc-frcnn_indices_%s' % split, feat_indices)
```
