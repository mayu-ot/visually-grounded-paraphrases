```python
from gensim.models import KeyedVectors
import pandas as pd
from gensim.parsing.preprocessing import remove_stopwords
from gensim.corpora import Dictionary
from gensim.parsing.preprocessing import preprocess_string
from gensim.parsing.preprocessing import strip_punctuation, strip_multiple_whitespaces, remove_stopwords, strip_numeric
import numpy as np

```

```python
cd /data/mayu-ot/Experiments/neurocomp_vgp/

```

```python
df = pd.read_csv('data/raw/train.csv',
                usecols=['original_phrase1',
                        'original_phrase2',
                        'ytrue'])
df.head()

```

```python
model = KeyedVectors.load_word2vec_format(
    '/data/mayu-ot/Data/Model/GoogleNews-vectors-negative300.bin.gz', binary=True)

```

```python
CUSTOM_FILTERS = [lambda x: x.lower(), strip_punctuation, strip_multiple_whitespaces, strip_numeric]
doc = [preprocess_string(x, CUSTOM_FILTERS) for x in df.values[:, :2].ravel()]
```

```python
dct = Dictionary(doc)

bad_ids = []
for k, v in dct.iteritems():
        if v not in model:
                bad_ids.append(k)
dct.filter_tokens(bad_ids)
```

```python
dct.compactify()

```

```python
for k, v in dct.iteritems():
    print(k, v)
    if k == 10:
        break

```

```python
dct.save_as_text('data/processed/dictionary_tmp.txt')

```

```python
word_emb = np.ones((len(dct), 300))

for k, v in dct.iteritems():
        word_emb[k] = model[v]

np.save('data/processed/word2vec', word_emb)

```
